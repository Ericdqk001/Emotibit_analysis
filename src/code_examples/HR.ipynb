```python
import pandas as pd

# === Set sensor type ===
sensor_type = 'HR'  # Change this to 'AX', 'AY', 'PG', etc.

# === Load EmotiBit sensor data for all subjects ===
df_SL = load_emotibit_sensor('./wrist/SL/2025-05-12_22-05-22-738924_' + sensor_type + '.csv', sensor_type=sensor_type)
df_FH = load_emotibit_sensor('./wrist/FH/2025-05-14_14-30-14-564821_' + sensor_type + '.csv', sensor_type=sensor_type)
df_SDLF = load_emotibit_sensor('./wrist/SDLF/2025-06-09_08-23-30-723495_' + sensor_type + '.csv', sensor_type=sensor_type)
df_Ali = load_emotibit_sensor('./wrist/Ali/2025-06-16_11-30-54-902836_' + sensor_type + '.csv', sensor_type=sensor_type)

# === Calculate stats for each subject ===
df_results_SL = calculate_task_stats(df_SL, 'task_times.txt', value_col=sensor_type)
df_results_FH = calculate_task_stats(df_FH, 'task_times.txt', value_col=sensor_type)
df_results_SDLF = calculate_task_stats(df_SDLF, 'task_times.txt', value_col=sensor_type)
df_results_Ali = calculate_task_stats(df_Ali, 'task_times.txt', value_col=sensor_type)

# === Combine and fill NaNs ===
all_dfs = [df_results_SL, df_results_FH, df_results_SDLF, df_results_Ali]
df_final = df_results_SL.copy()

mean_col = f'Mean_{sensor_type}'
std_col = f'STD_{sensor_type}'

for df_next in all_dfs[1:]:
    for idx, row in df_final.iterrows():
        if pd.isna(row[mean_col]) or row['Samples'] == 0:
            subject = row['Subject']
            task = row['Task']
            match = df_next[(df_next['Subject'] == subject) & (df_next['Task'] == task)]
            if not match.empty:
                match_row = match.iloc[0]
                if not pd.isna(match_row[mean_col]) and match_row['Samples'] > 0:
                    df_final.loc[idx, [mean_col, std_col, 'Samples']] = match_row[[mean_col, std_col, 'Samples']]

# === Final result ===
print(df_final)
```
